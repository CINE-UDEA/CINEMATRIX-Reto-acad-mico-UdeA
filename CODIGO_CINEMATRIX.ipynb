```python
# CINEMATRIX - Sistema de Cine Universitario

import os
import json
from datetime import datetime

# CONFIGURACION INICIAL

# Peliculas del fin de semana ------------------------------------ shadow
PELICULAS = {
    "1": {"nombre": "INTERSTELLAR", "hora": "2:00 PM", "dia": "VIERNES", "sala": "SALA A", "duracion": "169 min"},
    "2": {"nombre": "INTERSTELLAR", "hora": "4:30 PM", "dia": "VIERNES", "sala": "SALA A", "duracion": "169 min"},
    "3": {"nombre": "INTERSTELLAR", "hora": "7:00 PM", "dia": "VIERNES", "sala": "SALA A", "duracion": "169 min"},

    "4": {"nombre": "OPPENHEIMER", "hora": "2:00 PM", "dia": "SABADO", "sala": "SALA B", "duracion": "180 min"},
    "5": {"nombre": "OPPENHEIMER", "hora": "5:00 PM", "dia": "SABADO", "sala": "SALA B", "duracion": "180 min"},
    "6": {"nombre": "OPPENHEIMER", "hora": "8:00 PM", "dia": "SABADO", "sala": "SALA B", "duracion": "180 min"},

    "7": {"nombre": "THE IMITATION GAME", "hora": "2:00 PM", "dia": "DOMINGO", "sala": "SALA C", "duracion": "114 min"},
    "8": {"nombre": "THE IMITATION GAME", "hora": "4:30 PM", "dia": "DOMINGO", "sala": "SALA C", "duracion": "114 min"},
    "9": {"nombre": "THE IMITATION GAME", "hora": "7:00 PM", "dia": "DOMINGO", "sala": "SALA C", "duracion": "114 min"},

    "10": {"nombre": "ZOOTOPIA 2", "hora": "1:30 PM", "dia": "VIERNES", "sala": "SALA D", "duracion": "108 min"},
    "11": {"nombre": "ZOOTOPIA 2", "hora": "3:45 PM", "dia": "VIERNES", "sala": "SALA D", "duracion": "108 min"},
    "12": {"nombre": "ZOOTOPIA 2", "hora": "6:00 PM", "dia": "VIERNES", "sala": "SALA D", "duracion": "108 min"},

    "13": {"nombre": "EL SOBREVIVIENTE", "hora": "2:15 PM", "dia": "SABADO", "sala": "SALA E", "duracion": "109 min"},
    "14": {"nombre": "EL SOBREVIVIENTE", "hora": "4:45 PM", "dia": "SABADO", "sala": "SALA E", "duracion": "109 min"},
    "15": {"nombre": "EL SOBREVIVIENTE", "hora": "7:15 PM", "dia": "SABADO", "sala": "SALA E", "duracion": "109 min"},

    "16": {"nombre": "WICKED: POR SIEMPRE", "hora": "1:00 PM", "dia": "DOMINGO", "sala": "SALA F", "duracion": "137 min"},
    "17": {"nombre": "WICKED: POR SIEMPRE", "hora": "4:00 PM", "dia": "DOMINGO", "sala": "SALA F", "duracion": "137 min"},
    "18": {"nombre": "WICKED: POR SIEMPRE", "hora": "7:00 PM", "dia": "DOMINGO", "sala": "SALA F", "duracion": "137 min"},

    "19": {"nombre": "LOS ILUSIONISTAS 3", "hora": "2:30 PM", "dia": "VIERNES", "sala": "SALA G", "duracion": "113 min"},
    "20": {"nombre": "LOS ILUSIONISTAS 3", "hora": "5:00 PM", "dia": "VIERNES", "sala": "SALA G", "duracion": "113 min"},
    "21": {"nombre": "LOS ILUSIONISTAS 3", "hora": "7:30 PM", "dia": "VIERNES", "sala": "SALA G", "duracion": "113 min"},

    "22": {"nombre": "UNO VUELVE SIEMPRE", "hora": "3:00 PM", "dia": "SABADO", "sala": "SALA H", "duracion": "95 min"},
    "23": {"nombre": "UNO VUELVE SIEMPRE", "hora": "5:15 PM", "dia": "SABADO", "sala": "SALA H", "duracion": "95 min"},
    "24": {"nombre": "UNO VUELVE SIEMPRE", "hora": "7:30 PM", "dia": "SABADO", "sala": "SALA H", "duracion": "95 min"}
}

# Precios por tipo de usuario
PRECIOS = {
    "ESTUDIANTE": 7500,
    "DOCENTE": 10000,
    "ADMINISTRATIVO": 8500,
    "OFICIAL INTERNO": 7000,
    "PUBLICO EXTERNO": 15000
}

# Combos de confiteria
COMBOS = {
    "1": {"nombre": "COMBO 1", "descripcion": "Crispetas + 2 Gaseosas", "precio": 20000},
    "2": {"nombre": "COMBO 2", "descripcion": "Crispetas + 2 Gaseosas + 2 Perros", "precio": 60000},
    "3": {"nombre": "COMBO 3", "descripcion": "Crispetas + 1 Gaseosa", "precio": 14000},
    "4": {"nombre": "GASEOSA SOLA", "descripcion": "Gaseosa individual", "precio": 5000},
    "5": {"nombre": "PERRO SOLO", "descripcion": "Perro caliente individual", "precio": 8000}
}

#  FUNCIONES DE VISUALIZACION

def mostrar_logo():
    # Muestro el logo y nombre del cine
    print("\n" + "="*60)
    print("            CINEMATRIX")
    print("      DONDE CADA HISTORIA COBRA VIDA")
    print("="*60)

def dibujar_sala(asientos_ocupados):
    # Dibuja la sala de cine con 121 asientos distribuidos en 11x11
    print("\n" + "SALA DE CINE - DISTRIBUCION DE ASIENTOS")
    print("=" * 70)
    print("          PANTALLA PRINCIPAL")
    print("=" * 70)

    # Encabezado con numeros de columnas
    print("     ", end="")
    for i in range(1, 12):
        print(f"{i:2} ", end="")
    print()

    # Filas de asientos (A a K)
    for fila in range(11):
        letra_fila = chr(65 + fila)  # Convierte numero a letra (A, B, C, etc.)

        print(f" {letra_fila} |", end="")

        for columna in range(11):
            asiento = f"{letra_fila}{columna+1}"

            # Columnas 4 y 7 son pasillos
            if columna == 4 or columna == 7:
                print(" || ", end="")  # Simbolo de pasillo
            else:
                if asiento in asientos_ocupados:
                    print(" XX ", end="")  # Asiento ocupado
                else:
                    print(" OO ", end="")  # Asiento disponible

        print(" |")

        # Pasillo horizontal despues de la fila E
        if fila == 4:
            print("    " + "-" * 45 + " PASILLO CENTRAL " + "-" * 20)

    # Informacion de salidas
    print("\n" + "=" * 70)
    print("SALIDAS: IZQUIERDA - EN EL CENTRO - DERECHA")
    print("=" * 70)

    # Leyenda mejorada
    print("\nLEYENDA:")
    print("OO = Asiento disponible")
    print("XX = Asiento ocupado")
    print("|| = Pasillo")
    print("=" * 70)

# FUNCIONES DE VALIDACION

def validar_nombre_apellido(texto):
    # Valida que el nombre o el apellido tenga minimo 3 letras y no tenga numeros
    if len(texto) < 3:   # ejemplo ANA
        return False, "Error: Debe tener al menos 3 caracteres"
    if any(char.isdigit() for char in texto):
        return False, "Error: No se permiten numeros"
    return True, "Correcto"

def validar_documento(documento):
    # Valida que el documento tenga entre 6 y 10 digitos y solo numeros
    if not documento.isdigit():
        return False, "Error: Solo se permiten numeros"
    if len(documento) < 6 or len(documento) > 10:
        return False, "Error: Debe tener entre 6 y 10 digitos"
    return True, "Correcto"

def validar_correo_udea(correo):
    # Valida que el correo sea institucional de la UdeA
    if not correo.endswith('@udea.edu.co'):
        return False, "Error: Debe usar correo institucional @udea.edu.co"
    if len(correo) <= 12:  # @udea.edu.co tiene 12 caracteres
        return False, "Error: Correo muy corto"
    return True, "Correcto"

# -------xxxx-----xxx------ SISTEMA DE USUARIOS ----xxxx------xxxx---------

# Diccionarios para almacenar datos
usuarios_registrados = {}
usuario_actual = None

def registrar_usuario():
    # Registra un nuevo usuario en el sistema
    global usuarios_registrados

    mostrar_logo()
    print("\nREGISTRO DE NUEVO USUARIO")
    print("-" * 40)

    # Validar nombre
    while True:
        nombre = input("Ingrese SU nombre: ").strip().upper()
        es_valido, mensaje = validar_nombre_apellido(nombre)
        print(mensaje)
        if es_valido:
            break

    # Validar apellido
    while True:
        apellido = input("Ingrese SU apellido: ").strip().upper()
        es_valido, mensaje = validar_nombre_apellido(apellido)
        print(mensaje)
        if es_valido:
            break

    # Validar documento
    while True:
        documento = input("Ingrese SU documento de identificacion: ").strip()
        es_valido, mensaje = validar_documento(documento)
        print(mensaje)
        if es_valido:
            # Verificar si el documento ya existe
            if documento in usuarios_registrados:
                print("Error: Este documento ya se encuentra registrado")
                continue
            break

    # Validar correo institucional
    while True:
        correo = input("Ingrese su correo institucional (@udea.edu.co): ").strip().lower()
        es_valido, mensaje = validar_correo_udea(correo)
        print(mensaje)
        if es_valido:
            break

    # Seleccionar tipo de usuario -------------------------------------- shadow
    print("\nSELECCIONE SU TIPO DE USUARIO:")
    print("-" * 30)
    tipos = list(PRECIOS.keys())
    for i, tipo in enumerate(tipos, 1):
        print(f"{i}. {tipo} - ${PRECIOS[tipo]:,}")

    while True:
        try:
            opcion = int(input("Seleccione una opcion entre (1-5) donde debe solo escribir un numero: "))
            if 1 <= opcion <= 5:
                tipo_usuario = tipos[opcion-1]
                break
            else:
                print("Error: Opcion invalida. Seleccione 1-5")
        except ValueError:
            print("Error: Ingrese un numero valido")

    # Crear usuario y contrasena
    usuario = input("\nCree un nombre de usuario(RECUERDA NO OLVIDARLO): ").strip()
    contrasena = input("Cree una contrasena (RECUERDA NO OLVIDARLA): ").strip()

    # Guardar usuario en el diccionario
    usuarios_registrados[documento] = {
        "nombre": nombre,
        "apellido": apellido,
        "documento": documento,
        "correo": correo,
        "tipo_usuario": tipo_usuario,
        "precio_tiquete": PRECIOS[tipo_usuario],
        "usuario": usuario,
        "contrasena": contrasena,
        "reservas": [],
        "compras_confiteria": []
    }

    print(f"\n¡USUARIO REGISTRADO EXITOSAMENTE!")
    print(f"Bienvenido(a) {nombre} {apellido}")
    print(f"Tipo de usuario: {tipo_usuario}")
    print(f"Precio por tiquete: ${PRECIOS[tipo_usuario]:,}")
    input("\nPresione la tecla ENTER para continuar...")

def iniciar_sesion():
    # Permite iniciar sesion como usuario normal o administrador
    global usuario_actual

    mostrar_logo()
    print("\nINICIAR SESION")
    print("-" * 30)

    # Credenciales de administrador
    admin_usuario = "admin"
    admin_contrasena = "cine123"

    usuario = input("Usuario: ").strip()
    contrasena = input("Contrasena: ").strip()

    # Verificar si es administrador
    if usuario == admin_usuario and contrasena == admin_contrasena:
        usuario_actual = {"tipo": "ADMINISTRADOR", "usuario": usuario}
        print("\n¡BIENVENIDO ADMINISTRADOR!")
        return "admin"

    # Buscar usuario normal en el diccionario
    for doc, datos in usuarios_registrados.items():
        if datos["usuario"] == usuario and datos["contrasena"] == contrasena:
            usuario_actual = datos
            usuario_actual["tipo"] = "USUARIO"
            print(f"\n¡BIENVENIDO(A) {datos['nombre']}!")
            return "usuario"

    print("Error: Usuario o contrasena incorrectos")
    input("Presione ENTER para continuar...")
    return None

# SISTEMA DE RESERVAS

reservas = {}

def mostrar_cartelera():
    # Muestra todas las peliculas disponibles en cartelera
    mostrar_logo()
    print("\nCARTELERA DE ESTA SEMANA")
    print("=" * 80)

    for id_funcion, datos in PELICULAS.items():
        # Contar asientos ocupados para esta funcion
        asientos_ocupados = []
        if id_funcion in reservas:
            asientos_ocupados = list(reservas[id_funcion].keys())

        disponibles = 121 - len(asientos_ocupados)

        print(f"Funcion {id_funcion}. {datos['nombre']}")
        print(f"   Fecha: {datos['dia']} | Hora: {datos['hora']} | Sala: {datos['sala']}")
        print(f"   Duracion: {datos['duracion']} | Asientos disponibles: {disponibles}/121")
        print("-" * 80)

def hacer_reserva():
    # Permite a un usuario hacer una reserva de asiento --------------------------------------- shadow
    global reservas

    if usuario_actual["tipo"] != "USUARIO":
        print("Error: Solo los usuarios registrados pueden hacer reservas")
        input("Presione la tecla ENTER para continuar...")
        return

    mostrar_cartelera()

    # Seleccionar funcion
    while True:
        id_funcion = input("\nSeleccione el numero de la funcion entre (1-24): ").strip()
        if id_funcion not in PELICULAS:
            print("Error: Funcion no valida")
            continue
        break

    # Inicializar reservas para esta funcion si no existen ------------------------ shadow
    if id_funcion not in reservas:
        reservas[id_funcion] = {}

    # Mostrar sala con asientos ocupados
    asientos_ocupados = list(reservas[id_funcion].keys())
    dibujar_sala(asientos_ocupados)

    # Seleccionar asiento
    while True:
        asiento = input("\nSeleccione su asiento (ejemplo: A5, B3, etc.) escriba la letra en mayuscula: ").strip().upper()

        # Validar formato del asiento
        if len(asiento) < 2 or not asiento[0].isalpha() or not asiento[1:].isdigit():
            print("Error: Formato invalido. Use Letra seguida de Numero (ejemplo: A5)")
            continue

        letra = asiento[0]
        numero = int(asiento[1:])

        # Validar rango de asientos
        if letra < 'A' or letra > 'K' or numero < 1 or numero > 11:
            print("Error: Asiento no valido. Letras de A a K, Numeros de 1 a 11")
            continue

        # Verificar si el asiento esta ocupado
        if asiento in asientos_ocupados:
            print("Error: lo sentimos Este asiento ya esta ocupado")
            continue

        # Verificar si es un pasillo
        if numero == 5 or numero == 8:
            print("Error: Esta ubicacion es un pasillo, no es un asiento")
            continue

        break

    # Confirmar reserva
    funcion = PELICULAS[id_funcion]
    precio = usuario_actual["precio_tiquete"]

    print(f"\nCONFIRMACION DE RESERVA:")
    print(f"Pelicula: {funcion['nombre']}")
    print(f"Fecha y hora: {funcion['dia']} {funcion['hora']}")
    print(f"Sala: {funcion['sala']}")
    print(f"Asiento: {asiento}")
    print(f"Usuario: {usuario_actual['nombre']} {usuario_actual['apellido']}")
    print(f"Precio: ${precio:,}")

    confirmar = input("\n¿Desea confirmar la reserva?, escribra entre (SI/NO) y luego precione la tecla ENTER: ").strip().upper()

    if confirmar == "SI" or confirmar == "S":
        # Guardar reserva en el diccionario
        reservas[id_funcion][asiento] = {
            "documento": usuario_actual["documento"],
            "nombre": usuario_actual["nombre"] + " " + usuario_actual["apellido"],
            "tipo_usuario": usuario_actual["tipo_usuario"],
            "precio": precio,
            "fecha": datetime.now().strftime("%Y-%m-%d %H:%M")
        }

        # Agregar a historial del usuario
        usuario_actual["reservas"].append({
            "funcion": id_funcion,
            "asiento": asiento,
            "pelicula": funcion['nombre'],
            "fecha_funcion": f"{funcion['dia']} {funcion['hora']}",
            "precio": precio,
            "estado": "ACTIVA"
        })

        print("\n¡RESERVA CONFIRMADA EXITOSAMENTE!")
        generar_factura(id_funcion, asiento, precio)
    else:
        print("Reserva cancelada")

    input("\nPresione la tecla ENTER para continuar...")

def cancelar_reserva():
    # Permite a un usuario cancelar una reserva activa
    global reservas

    if usuario_actual["tipo"] != "USUARIO":
        print("Error: Solo los usuarios registrados pueden cancelar reservas")
        input("Presione la tecla ENTER para continuar...")
        return

    # Verificar si el usuario tiene reservas
    if not usuario_actual["reservas"]:
        print("No tiene reservas activas para cancelar")
        input("Presione la tecla ENTER para continuar...")
        return

    # Mostrar reservas activas del usuario
    print("\nMIS RESERVAS ACTIVAS:")
    print("=" * 80)

    reservas_activas = []
    for i, reserva in enumerate(usuario_actual["reservas"]):
        if reserva.get("estado", "ACTIVA") == "ACTIVA":
            reservas_activas.append((i, reserva))
            print(f"{len(reservas_activas)}. Pelicula: {reserva['pelicula']}")
            print(f"   Funcion: {reserva['fecha_funcion']}")
            print(f"   Asiento: {reserva['asiento']}")
            print(f"   Precio: ${reserva['precio']:,}")
            print("-" * 80)

    if not reservas_activas:
        print("No tiene reservas activas")
        input("Presione la tecla ENTER para continuar...")
        return

    # Seleccionar reserva a cancelar
    while True:
        try:
            opcion = int(input(f"\nSeleccione el numero de reserva a cancelar (1-{len(reservas_activas)}): "))
            if 1 <= opcion <= len(reservas_activas):
                reserva_seleccionada = reservas_activas[opcion-1]
                break
            else:
                print(f"Error: Seleccione un numero entre 1 y {len(reservas_activas)}")
        except ValueError:
            print("Error: Ingrese un numero valido")

    indice_original, reserva = reserva_seleccionada
    id_funcion = reserva["funcion"]
    asiento = reserva["asiento"]

    # Confirmar cancelacion
    print(f"\nCONFIRMACION DE CANCELACION:")
    print(f"Pelicula: {reserva['pelicula']}")
    print(f"Funcion: {reserva['fecha_funcion']}")
    print(f"Asiento: {asiento}")
    print(f"Precio a reembolsar: ${reserva['precio']:,}")

    confirmar = input("\n¿Esta seguro de que desea cancelar esta reserva?, escriba en mayuscula la palabra (SI/NO) y luego precione la tecla ENTER: ").strip().upper()

    if confirmar == "SI" or confirmar == "S":
        # Eliminar reserva del sistema general
        if id_funcion in reservas and asiento in reservas[id_funcion]:
            del reservas[id_funcion][asiento]

        # Marcar reserva como cancelada en el historial del usuario
        usuario_actual["reservas"][indice_original]["estado"] = "CANCELADA"
        usuario_actual["reservas"][indice_original]["fecha_cancelacion"] = datetime.now().strftime("%Y-%m-%d %H:%M")

        print("\n¡RESERVA CANCELADA EXITOSAMENTE!")
        print(f"Se ha liberado el asiento {asiento}")
        print(f"Se ha echo el procesado de reembolso de ${reserva['precio']:,}")
    else:
        print("Cancelacion abortada")

    input("\nPresione la tecla ENTER para continuar...")

def generar_factura(id_funcion, asiento, precio):
    # Genera una factura de la reserva realizada
    funcion = PELICULAS[id_funcion]

    print("\n" + "=" * 50)
    print("           FACTURA CINEMATRIX")
    print("=" * 50)
    print(f"Cliente: {usuario_actual['nombre']} {usuario_actual['apellido']}")
    print(f"Documento: {usuario_actual['documento']}")
    print(f"Tipo de usuario: {usuario_actual['tipo_usuario']}")
    print("-" * 50)
    print(f"Pelicula: {funcion['nombre']}")
    print(f"Fecha: {funcion['dia']}")
    print(f"Hora: {funcion['hora']}")
    print(f"Sala: {funcion['sala']}")
    print(f"Asiento: {asiento}")
    print(f"Duracion: {funcion['duracion']}")
    print("-" * 50)
    print(f"TOTAL A PAGAR: ${precio:,}")
    print("=" * 50)
    print("¡Gracias por su compra! Disfrute la funcion")
    print("=" * 50)

# -------00000-----------SISTEMA DE CONFITERIA---------------000---------

def menu_confiteria():
    # Muestra el menu de confiteria y procesa compras
    while True:
        mostrar_logo()
        print("\nCONFITERIA - MENU DE COMIDAS Y BEBIDAS")
        print("=" * 50)
        print("1. COMBO 1 - Crispetas + 2 Gaseosas - $20,000")
        print("2. COMBO 2 - Crispetas + 2 Gaseosas + 2 Perros - $60,000")
        print("3. COMBO 3 - Crispetas + 1 Gaseosa - $14,000")
        print("4. GASEOSA SOLA - $5,000")
        print("5. PERRO SOLO - $8,000")
        print("6. VER MIS COMPRAS DE CONFITERIA")
        print("7. CANCELAR COMPRA DE CONFITERIA")
        print("8. VOLVER AL MENU PRINCIPAL")
        print("=" * 50)

        opcion = input("Seleccione una opcion (1-8): ").strip()

        if opcion == "8":
            break
        elif opcion == "6":
            ver_compras_confiteria()
        elif opcion == "7":
            cancelar_compra_confiteria()
        elif opcion in COMBOS:
            realizar_compra_confiteria(opcion)
        else:
            print("Error: Opcion no valida")
            input("Presione ENTER para continuar...")

def realizar_compra_confiteria(opcion_combo):
    # Procesa una compra de confiteria
    combo = COMBOS[opcion_combo]
    print(f"\nHa seleccionado: {combo['nombre']}")
    print(f"Descripcion: {combo['descripcion']}")
    print(f"Precio: ${combo['precio']:,}")

    confirmar = input("¿Confirmar compra? (SI/NO): ").strip().upper()

    if confirmar == "SI" or confirmar == "S":
        # Guardar compra en el historial del usuario ------------------------------ shadow
        if "compras_confiteria" not in usuario_actual:
            usuario_actual["compras_confiteria"] = []

        compra = {
            "combo": combo['nombre'],
            "descripcion": combo['descripcion'],
            "precio": combo['precio'],
            "fecha": datetime.now().strftime("%Y-%m-%d %H:%M"),
            "estado": "ACTIVA"
        }

        usuario_actual["compras_confiteria"].append(compra)

        print(f"\n¡COMPRA CONFIRMADA!")
        print(f"Producto: {combo['nombre']}")
        print(f"Total: ${combo['precio']:,}")
        input("Presione la tecla ENTER para continuar...")
    else:
        print("Compra cancelada")
        input("Presione la tecla ENTER para continuar...")

def ver_compras_confiteria():
    # Muestra el historial de compras de confiteria del usuario
    print("\nMIS COMPRAS DE CONFITERIA")
    print("=" * 60)

    if not usuario_actual.get("compras_confiteria"):
        print("No tiene compras de confiteria registradas")
    else:
        compras_activas = [c for c in usuario_actual["compras_confiteria"] if c.get("estado", "ACTIVA") == "ACTIVA"]

        if not compras_activas:
            print("No tiene compras activas de confiteria")
        else:
            for i, compra in enumerate(compras_activas, 1):
                print(f"{i}. {compra['combo']}")
                print(f"   Descripcion: {compra['descripcion']}")
                print(f"   Precio: ${compra['precio']:,}")
                print(f"   Fecha: {compra['fecha']}")
                print("-" * 60)

    input("\nPresione la tecla ENTER para continuar...")

def cancelar_compra_confiteria():
    # Permite cancelar una compra de confiteria
    print("\nCANCELAR COMPRA DE CONFITERIA")
    print("=" * 60)

    if not usuario_actual.get("compras_confiteria"):
        print("No tiene compras de confiteria para cancelar")
        input("Presione la tecla ENTER para continuar...")
        return

    compras_activas = [c for c in usuario_actual["compras_confiteria"] if c.get("estado", "ACTIVA") == "ACTIVA"]

    if not compras_activas:
        print("No tiene compras activas de confiteria")
        input("Presione la tecla ENTER para continuar...")
        return

    # Mostrar compras activas
    for i, compra in enumerate(compras_activas, 1):
        print(f"{i}. {compra['combo']} - ${compra['precio']:,} - {compra['fecha']}")

    # Seleccionar compra a cancelar
    while True:
        try:
            opcion = int(input(f"\nSeleccione el numero de compra a cancelar (1-{len(compras_activas)}): "))
            if 1 <= opcion <= len(compras_activas):
                compra_seleccionada = compras_activas[opcion-1]
                break
            else:
                print(f"Error: Seleccione un numero entre 1 y {len(compras_activas)}")
        except ValueError:
            print("Error: Ingrese un numero valido")

    # Confirmar cancelacion
    print(f"\nCONFIRMACION DE CANCELACION:")
    print(f"Producto: {compra_seleccionada['combo']}")
    print(f"Descripcion: {compra_seleccionada['descripcion']}")
    print(f"Precio a reembolsar: ${compra_seleccionada['precio']:,}")

    confirmar = input("\n¿Esta seguro de que desea cancelar esta compra? (SI/NO): ").strip().upper()

    if confirmar == "SI" or confirmar == "S":
        # Encontrar y marcar la compra como cancelada
        for compra in usuario_actual["compras_confiteria"]:
            if (compra.get("estado", "ACTIVA") == "ACTIVA" and
                compra["combo"] == compra_seleccionada["combo"] and
                compra["fecha"] == compra_seleccionada["fecha"]):
                compra["estado"] = "CANCELADA"
                compra["fecha_cancelacion"] = datetime.now().strftime("%Y-%m-%d %H:%M")
                break

        print("\n¡COMPRA CANCELADA EXITOSAMENTE!")
        print(f"Se ha echo el procesado del reembolso de ${compra_seleccionada['precio']:,}")
    else:
        print("Cancelacion abortada")

    input("\nPresione ENTER para continuar...")

# ----------------xxx---------------- MODULO DE ADMINISTRADOR ---------xxx--------------shadow

def panel_administrador():
    # Panel de control exclusivo para administradores
    while True:
        mostrar_logo()
        print("\nPANEL DE ADMINISTRADOR")
        print("=" * 50)
        print("1. Ver reportes generales")
        print("2. Ver estado de funciones")
        print("3. Ver lista de usuarios")
        print("4. Ver detalles de sala especifica")
        print("5. Volver al menu principal")

        opcion = input("\nSeleccione una opcion: ").strip()

        if opcion == "1":
            ver_reportes_generales()
        elif opcion == "2":
            ver_estado_funciones()
        elif opcion == "3":
            ver_lista_usuarios()
        elif opcion == "4":
            ver_detalles_sala()
        elif opcion == "5":
            break
        else:
            print("Error: Opcion no valida")
            input("Presione ENTER para continuar...")

def ver_reportes_generales():
    # Muestra reportes generales del sistema
    print("\nREPORTES GENERALES DEL SISTEMA")
    print("=" * 60)

    total_reservas = 0
    total_ingresos = 0
    reservas_por_tipo = {}

    # Calcular totales
    for id_funcion, asientos in reservas.items():
        for asiento, datos in asientos.items():
            total_reservas += 1
            total_ingresos += datos["precio"]
            tipo = datos["tipo_usuario"]
            reservas_por_tipo[tipo] = reservas_por_tipo.get(tipo, 0) + 1

    print(f"Total de reservas: {total_reservas}")
    print(f"Total de ingresos: ${total_ingresos:,}")
    print(f"Total de usuarios registrados: {len(usuarios_registrados)}")

    print("\nReservas por tipo de usuario:")
    for tipo, cantidad in reservas_por_tipo.items():
        print(f"   {tipo}: {cantidad} reservas")

    print("\nEstado de funciones:")
    for id_funcion, datos in PELICULAS.items():
        ocupados = len(reservas.get(id_funcion, {}))
        disponibles = 121 - ocupados
        print(f"   Funcion {id_funcion}. {datos['nombre']}: {ocupados}/121 asientos ocupados")

    input("\nPresione ENTER para continuar...")

def ver_estado_funciones():
    # Muestra el estado detallado de todas las funciones
    print("\nESTADO DETALLADO DE FUNCIONES")
    print("=" * 80)

    for id_funcion, datos in PELICULAS.items():
        asientos_funcion = reservas.get(id_funcion, {})
        ocupados = len(asientos_funcion)
        ingresos_funcion = sum(datos_reserva["precio"] for datos_reserva in asientos_funcion.values())

        print(f"\nFuncion {id_funcion}. {datos['nombre']}")
        print(f"   Fecha: {datos['dia']} | Hora: {datos['hora']} | Sala: {datos['sala']}")
        print(f"   Asientos ocupados: {ocupados}/121")
        print(f"   Ingresos: ${ingresos_funcion:,}")

        if ocupados > 0:
            print("   Usuarios en esta funcion:")
            for asiento, datos_reserva in asientos_funcion.items():
                print(f"      - {datos_reserva['nombre']} (Asiento {asiento})")

    input("\nPresione ENTER para continuar...")

def ver_lista_usuarios():
    # Muestra la lista de todos los usuarios registrados
    print("\nLISTA DE USUARIOS REGISTRADOS")
    print("=" * 60)

    if not usuarios_registrados:
        print("No hay usuarios registrados en el sistema")
    else:
        for i, (doc, datos) in enumerate(usuarios_registrados.items(), 1):
            print(f"\n{i}. {datos['nombre']} {datos['apellido']}")
            print(f"   Documento: {doc}")
            print(f"   Correo: {datos['correo']}")
            print(f"   Tipo de usuario: {datos['tipo_usuario']}")
            print(f"   Precio tiquete: ${datos['precio_tiquete']:,}")
            print(f"   Usuario: {datos['usuario']}")
            print(f"   Reservas activas: {len([r for r in datos['reservas'] if r.get('estado', 'ACTIVA') == 'ACTIVA'])}")
            print(f"   Compras confiteria activas: {len([c for c in datos.get('compras_confiteria', []) if c.get('estado', 'ACTIVA') == 'ACTIVA'])}")

    input("\nPresione ENTER para continuar...")

def ver_detalles_sala():
    # Permite al administrador ver detalles especificos de una sala
    mostrar_cartelera()

    id_funcion = input("\nIngrese el numero de la funcion para ver detalles: ").strip()

    if id_funcion not in PELICULAS:
        print("Error: Funcion no valida")
        input("Presione ENTER para continuar...")
        return

    asientos_ocupados = list(reservas.get(id_funcion, {}).keys())
    dibujar_sala(asientos_ocupados)

    print(f"\nDETALLES DE LA FUNCION {id_funcion}:")
    print(f"Pelicula: {PELICULAS[id_funcion]['nombre']}")
    print(f"Asientos ocupados: {len(asientos_ocupados)}")

    if asientos_ocupados:
        print("Detalle de ocupacion:")
        for asiento in asientos_ocupados:
            datos_reserva = reservas[id_funcion][asiento]
            print(f"   - Asiento {asiento}: {datos_reserva['nombre']} ({datos_reserva['tipo_usuario']})")

    input("\nPresione ENTER para continuar...")

# --------------------- MENU PRINCIPAL --------------
#shadow
def menu_principal():
    # Menu principal del sistema - controla el flujo de navegacion
    global usuario_actual

    while True:
        mostrar_logo()
        print("\nMENU PRINCIPAL")
        print("=" * 40)

        if usuario_actual is None:
            # Menu para usuarios no logueados
            print("1. Registrarse como nuevo usuario")
            print("2. Iniciar sesion")
            print("3. Ver cartelera (sin registro)")
            print("4. Salir del sistema")
        elif usuario_actual["tipo"] == "USUARIO":
            # Menu para usuarios normales
            print(f"Bienvenido: {usuario_actual['nombre']} {usuario_actual['apellido']}")
            print("1. Ver cartelera")
            print("2. Hacer reserva")
            print("3. Cancelar reserva")
            print("4. Confiteria")
            print("5. Ver mis reservas")
            print("6. Cerrar sesion")
        elif usuario_actual["tipo"] == "ADMINISTRADOR":
            # Menu para administradores
            print("Modo: ADMINISTRADOR")
            print("1. Ver cartelera")
            print("2. Panel de administrador")
            print("3. Cerrar sesion")

        print("=" * 40)

        opcion = input("Seleccione una opcion: ").strip()

        if usuario_actual is None:
            if opcion == "1":
                registrar_usuario()
            elif opcion == "2":
                iniciar_sesion()
            elif opcion == "3":
                mostrar_cartelera()
                input("\nPresione ENTER para continuar...")
            elif opcion == "4":
                print("\n¡Gracias por visitar CINEMATRIX!")
                print("Sistema cerrado.")
                break
            else:
                print("Error: Opcion no valida")
                input("Presione ENTER para continuar...")

        elif usuario_actual["tipo"] == "USUARIO":
            if opcion == "1":
                mostrar_cartelera()
                input("\nPresione ENTER para continuar...")
            elif opcion == "2":
                hacer_reserva()
            elif opcion == "3":
                cancelar_reserva()
            elif opcion == "4":
                menu_confiteria()
            elif opcion == "5":
                print("\nFuncion en desarrollo - Proximamente...")
                input("Presione ENTER para continuar...")
            elif opcion == "6":
                usuario_actual = None
                print("\nSesion cerrada exitosamente")
                input("Presione ENTER para continuar...")
            else:
                print("Error: Opcion no valida")
                input("Presione ENTER para continuar...")

        elif usuario_actual["tipo"] == "ADMINISTRADOR":
            if opcion == "1":
                mostrar_cartelera()
                input("\nPresione ENTER para continuar...")
            elif opcion == "2":
                panel_administrador()
            elif opcion == "3":
                usuario_actual = None
                print("\nSesion de administrador cerrada")
                input("Presione ENTER para continuar...")
            else:
                print("Error: Opcion no valida")
                input("Presione ENTER para continuar...")

#  -----------------------INICIO DEL PROGRAMA -------------------xx------------------- shadow

if __name__ == "__main__":
    # Punto de entrada principal del programa

    # Datos de ejemplo para pruebas
    usuarios_registrados = {
        "123456789": {
            "nombre": "ANA",
            "apellido": "GOMEZ",
            "documento": "123456789",
            "correo": "ana.gomez@udea.edu.co",
            "tipo_usuario": "ESTUDIANTE",
            "precio_tiquete": 7500,
            "usuario": "ana123",
            "contrasena": "clave123",
            "reservas": [],
            "compras_confiteria": []
        }
    }

    reservas = {
        "1": {"A5": {"documento": "123456789", "nombre": "ANA GOMEZ", "tipo_usuario": "ESTUDIANTE", "precio": 7500, "fecha": "2024-03-21 10:00"}},
        "10": {"B3": {"documento": "123456789", "nombre": "ANA GOMEZ", "tipo_usuario": "ESTUDIANTE", "precio": 7500, "fecha": "2024-03-21 10:30"}}
    }

    print("¡Bienvenido a CINEMATRIX!")
    print("Sistema de gestion cinematografica universitaria")
    input("\nPresione la tecla ENTER para comenzar...")

    # Iniciar el menu principal
    menu_principal()
```
